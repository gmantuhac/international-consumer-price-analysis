```python
import mysql.connector as sql
import numpy as np
import matplotlib.pyplot as plt

# Preparing to connect to MySQL server in order to extract my data
mydb = sql.connect (
    host="localhost",
    user="root",
    passwd="Mantuhac1*",
    database='third',
    use_pure=True
)

print(mydb)
```

```python
# Test to see whether I can execute commands and extract the data from my tables
mycursor = mydb.cursor()
mycursor.execute("SELECT * FROM switzerland_prices")
result = mycursor.fetchall()

# Sanity check to see whether I successfully connected to server and can extract data from tables
for x in result:
    print(x)
```

```python
mycursor.execute("SELECT EndDate, Value from china_prices")
result_cn = mycursor.fetchall()

# Creating a label list, a measurable data list, and a count list
end_dates_xticks_cn = []
consumer_prices_cn = []
x_cn = []

# Inserting data from consumer prices in China
count = 0
for i in result_cn:
    end_dates_xticks_cn.append(i[0])
    consumer_prices_cn.append(float(i[1]))
    x_cn.append(count)
    count += 1
    
print(end_dates_xticks_cn)
print(consumer_prices_cn)
print(x_cn)
```

```python
mycursor.execute("SELECT EndDate, Value from dominica_prices")
result_dm = mycursor.fetchall()

# Creating a label list, a measurable data list, and a count list
end_dates_xticks_dm = []
consumer_prices_dm = []
x_dm = []

# Inserting data from consumer prices in Dominica
count = 0
for j in result_dm:
    end_dates_xticks_dm.append(j[0])
    consumer_prices_dm.append(float(j[1]))
    x_dm.append(count)
    count += 1
    
print(end_dates_xticks_dm)
print(consumer_prices_dm)
print(x_dm)
```

```python
mycursor.execute("SELECT EndDate, Value from lithuania_prices")
result_lt = mycursor.fetchall()

# Creating a label list, a measurable data list, and a count list
end_dates_xticks_lt = []
consumer_prices_lt = []
x_lt = []

# Inserting data from consumer prices in Lithuania
count = 0
for k in result_lt:
    end_dates_xticks_lt.append(k[0])
    consumer_prices_lt.append(float(k[1]))
    x_lt.append(count)
    count += 1
    
print(end_dates_xticks_lt)
print(consumer_prices_lt)
print(x_lt)
```

```python
mycursor.execute("SELECT EndDate, Value from switzerland_prices")
result_ch = mycursor.fetchall()

# Creating a label list, a measurable data list, and a count list
end_dates_xticks_ch = []
consumer_prices_ch = []
x_ch = []

# Inserting data from consumer prices in Switzerland 
# Note: CH is the ISO country code for Switzerland, to avoid any confusion
count = 0
for l in result_ch:
    end_dates_xticks_ch.append(l[0])
    consumer_prices_ch.append(float(l[1]))
    x_ch.append(count)
    count += 1
    
print(end_dates_xticks_ch)
print(consumer_prices_ch)
print(x_ch)
```

```python
# Preparing a regression line to mark the general trend
coefficients_cn = np.polyfit(x_cn, consumer_prices_cn, 1)
line_cn = np.poly1d(coefficients_cn)

# Plotting the data points from China
plt.plot(x_cn, consumer_prices_cn, label='Prices')
plt.plot(x_cn, line_cn(x_cn), color='red', label='Consumer Price Trend')
plt.xlabel('Months')
plt.ylabel('Consumer Prices')
plt.title('China Consumer Prices 01/2000 - 12/2022', loc='right')
plt.legend()
plt.show()
```

```python
# Preparing a regression line to mark the general trend
coefficients_dm = np.polyfit(x_dm, consumer_prices_dm, 1)
line_dm = np.poly1d(coefficients_dm)

# Plotting the data points from Dominica
plt.plot(x_dm, consumer_prices_dm, label='Prices')
plt.plot(x_dm, line_dm(x_dm), color='red', label='Consumer Price Trend')
plt.xlabel('Months')
plt.ylabel('Consumer Prices')
plt.title('Dominica Consumer Prices 01/2000 - 12/2022', loc='right')
plt.legend()
plt.show()
```

```python
# Preparing a regression line to mark the general trend
coefficients_lt = np.polyfit(x_lt, consumer_prices_lt, 1)
line_lt = np.poly1d(coefficients_lt)

# Plotting the data points from Lithuania
plt.plot(x_lt, consumer_prices_lt, label='Prices')
plt.plot(x_lt, line_lt(x_lt), color='red', label='Consumer Price Trend')
plt.xlabel('Months')
plt.ylabel('Consumer Prices')
plt.title('Lithuania Consumer Prices 01/2000 - 12/2022', loc='right')
plt.legend()
plt.show()
```

```python
# Preparing a regression line to mark the general trend
coefficients_ch = np.polyfit(x_ch, consumer_prices_ch, 1)
line_ch = np.poly1d(coefficients_ch)

# Plotting the data points from Switzerland
plt.plot(x_ch, consumer_prices_ch, label='Prices')
plt.plot(x_ch, line_ch(x_ch), color='red', label='Consumer Price Trend')
plt.xlabel('Months')
plt.ylabel('Consumer Prices')
plt.title('Switzerland Consumer Prices 01/2000 - 12/2022', loc='right')
plt.legend()
plt.show()
```

```python
# Plotting a subplot to compare each country's consumer price points using shared axis
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
axs[0, 0].plot(x_cn, consumer_prices_cn, color='purple', label='Prices')
axs[0, 0].plot(x_cn, line_cn(x_cn), color='red', label='Price Trend')
axs[0, 0].legend(prop={'size': 6}, loc='upper left')
axs[0, 0].set_title('China')
axs[0, 1].plot(x_dm, consumer_prices_dm, color='green', label='Prices')
axs[0, 1].plot(x_dm, line_dm(x_dm), color='red', label='Price Trend')
axs[0, 1].legend(prop={'size': 6}, loc='upper left')
axs[0, 1].set_title('Dominica')
axs[1, 0].plot(x_lt, consumer_prices_lt, color='blue', label='Prices')
axs[1, 0].plot(x_lt, line_lt(x_lt), color='red', label='Price Trend')
axs[1, 0].legend(prop={'size': 6}, loc='upper left')
axs[1, 0].set_title('Lithuania')
axs[1, 1].plot(x_ch, consumer_prices_ch, color='orange', label='Prices')
axs[1, 1].plot(x_ch, line_ch(x_ch), color='red', label='Price Trend')
axs[1, 1].legend(prop={'size': 6}, loc='upper left')
axs[1, 1].set_title('Switzerland')

fig.suptitle('Consumer Prices 01/2000 - 12/2022', fontsize=9)

# Setting the axis for each plot
for ax in axs.flat:
    ax.set(xlabel='Months', ylabel='Consumer Prices')

# Removing the axis labeels for top left and top right plots
for ax in axs.flat:
    ax.label_outer()
```
